# Identify outliers in price columns

```python
sns.boxplot(data=df,x="price")
```

```python
plt.figure(figsize = (15,6))
sns.distplot(df['price'], color = 'red', hist_kws={"linewidth": 15,'alpha':1})
plt.title("Probability Distribution", fontsize = 15)
plt.xlabel('Price', fontsize = 15,fontweight='bold')
plt.ylabel('Density', fontsize = 15,fontweight='bold')
plt.show()
```

```python
# calculating interquartile range
q25,q75=np.percentile(df["price"],[25,75])
IQR=q75-q25
q0=q25-(1.5*IQR)
q100=q75+(1.5*IQR)
print("IQR",IQR)
print("Minumum value: ", q0)
print("Maximum value: ", q100)
```

# Create new dataframe of price with range 0 to 377 this range is not outlier 
as showing above boxplot distribution and calcualted interquertile range of price is between the range of 0 to 377.

```python
df_new=df[df["price"]<377]
```

```python
df_new["price"].describe().reset_index().T
```

# Plot histogram to see price distribution on new data frame price

```python
plt.figure(figsize=(16,7))
plt.title("Price Distribution",fontweight='bold')
p=sns.histplot(data=df_new,x="price",kde=True,bins=100)
plt.ylabel("Frequency",fontweight='bold')
plt.xlabel("Price",fontweight='bold')
plt.show()
```

># Observation:
1. The distribution is highly skewed to the right, meaning there are a few very high-priced properties,
but most properties are priced relatively low.
2. The majority of prices are between 0 and 377 with mean 132 and median 119.
3. There is a long tail in the distribution, indicating the presence of some outliers with very high prices.

# Conclusion

Based on the observations from the Airbnb dataset analysis, several key insights can be drawn:

### Pricing and Distribution:
* **Price Skewness:** The distribution of property prices is heavily skewed to the right, indicating a significant number of lower-priced properties with fewer, more expensive outliers.
* **Neighborhood Variation:** Manhattan is the most expensive neighborhood, followed by Brooklyn, while Queens, Staten Island, and the Bronx exhibit more affordable pricing.
* **Room Type Pricing:** Entire home/apartments are generally more expensive than private rooms, hotel rooms, and shared rooms, across all neighborhoods.

### Reviews and Availability:
* **Positive Correlation:** There's a positive correlation between the number of reviews and property prices, suggesting that properties with more reviews tend to be priced higher.
* **Availability Impact:** Properties with higher availability, especially entire homes/apartments, often have lower prices.
* **Review Distribution:** Most reviews fall within the 0-400 price range, indicating a concentration of reviews for mid-priced properties.

### Geographical Distribution:
* **Listing Density:** Manhattan has the highest density of listings, followed by Brooklyn. Staten Island and the Bronx have fewer listings.
* **Room Type Clustering:** Entire home/apartments are more evenly distributed, while private rooms and hotel rooms have higher concentrations in certain areas, particularly Manhattan.

### Host Analysis:
* **Host Popularity:** The Ridge is the busiest host, followed by M, with private rooms being the most popular room type among top hosts.
* **Listing Volume:** Host 162280872 has the most listings, followed by host 51501835.

### Overall Trends:
* **Room Type Preferences:** Entire home/apartments are the most popular room type, followed by private rooms. Shared rooms are the least popular.
* **Neighborhood Factors:** Neighborhood significantly influences property prices, with Manhattan being the most expensive and Staten Island being the least expensive.
* **Pricing Dynamics:** The interplay of price, minimum nights, number of reviews, and availability varies across room types and neighborhoods.

In conclusion, the Airbnb dataset analysis reveals valuable insights into pricing trends, room type preferences, geographical distribution, and host activity. These findings can be utilized by potential hosts, travelers, and market analysts to make informed decisions and understand the dynamics of the Airbnb market.
